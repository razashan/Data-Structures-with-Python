```python
def bubble_sort(lis):
    for i in range(len(lis)):
        for j in range(0,len(lis)-i-1):
            if lis[j] > lis[j+1]:
                lis[j],lis[j+1] = lis[j+1],lis[j]
    return lis
```

```python
arr = [12,434,22,442]
ar = bubble_sort(arr)
```
