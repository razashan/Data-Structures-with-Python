```python
a = [12,3,45,32,43]
min_elem = 0
for i in range(len(a)):
    min_elem = i
    for j in range(i+1,len(a)):
        if a[min_elem] >a[j]:
            min_elem = j
    a[i],a[min_elem]= a[min_elem],a[i]

print("SORTED ARRAY")
for i in range(len(a)):
    print(a[i])
```
