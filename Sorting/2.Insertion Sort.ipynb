```python
def insertionsort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and key<arr[j]:
            arr[j+1] = arr[j]
            j -=1
            
        arr[j+1] =key
    return arr
```

```python
ls = [12,11,22,44]
print(insertionsort(ls))
```
